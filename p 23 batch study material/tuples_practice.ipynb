### Question 8
Create a tuple and **count** how many times an element appears.

**Input:** `(1, 2, 2, 3)`, check `2`  
**Expected Output:** `Count = 2`

````python
# Your solution here
t = (1,2,2,3)
count = 0
for i in t:
    if i == 2:
        count += 1

print(count)
````

### Question 9
Create a tuple and find the **index** of an element.

**Input:** `(10, 20, 30)`, find `20`  
**Expected Output:** `Index = 1`

````python
# Your solution here
t = (10,20,30)
for i in range(len(t)):
    if t[i] == 20:
        print(i)
````
